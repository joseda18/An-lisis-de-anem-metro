```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Caso_4._Energia_Eolica.xlsx", sheet_name="Datos de viento")
```

```python
df["Date"] = pd.to_datetime(df["Date"], format='%Y.%m.%d')
```

### Division por años

```python
a2008 = df[df["Date"].dt.year == 2008]
a2009 = df[df["Date"].dt.year == 2009]
```

### Promedio de Velocidad por mes:

```python
meses1 =  ['Marzo 2008', 'Abril 2008', 'Mayo 2008', 'Junio 2008', 'Julio 2008', 'Agosto 2008', 'Setiembre 2008', 'Octubre 2008', 'Noviembre 2008', 'Diciembre 2008']
meses2 =  ['Enero 2009', 'Febrero 2009', 'Marzo 2009', 'Abril 2009', 'Mayo 2009']

Promedios = {}

for i,mes1 in enumerate(meses1):
    Promedios[f'{mes1}'] = a2008[a2008["Date"].dt.month == i+3]["C2-Anemómetro 2 [m/s]"].mean()

for j,mes2 in enumerate(meses2):
    Promedios[f'{mes2}'] = a2009[a2009["Date"].dt.month == j+1]["C2-Anemómetro 2 [m/s]"].mean()
    
Promedios = pd.DataFrame([Promedios])
Promedios
```

### Maximos Por mes 

```python
Maximos = {}

for i,mes1 in enumerate(meses1):
    Maximos[f'{mes1}'] = a2008[a2008["Date"].dt.month == i+3]["C2-Anemómetro 2 [m/s]"].max()

for j,mes2 in enumerate(meses2):
    Maximos[f'{mes2}'] = a2009[a2009["Date"].dt.month == j+1]["C2-Anemómetro 2 [m/s]"].max()
    
Maximos = pd.DataFrame([Maximos])
Maximos
```

### Minimos por mes

```python
Minimos = {}

for i,mes1 in enumerate(meses1):
    Minimos[f'{mes1}'] = a2008[a2008["Date"].dt.month == i+3]["C2-Anemómetro 2 [m/s]"].min()

for j,mes2 in enumerate(meses2):
    Minimos[f'{mes2}'] = a2009[a2009["Date"].dt.month == j+1]["C2-Anemómetro 2 [m/s]"].min()
    
Minimos = pd.DataFrame([Minimos])
Minimos
```

### Desviacion Estandar

```python
Desv = {}

for i,mes1 in enumerate(meses1):
    Desv[f'{mes1}'] = a2008[a2008["Date"].dt.month == i+3]["C2-Anemómetro 2 [m/s]"].std(ddof =1)

for j,mes2 in enumerate(meses2):
    Desv[f'{mes2}'] = a2009[a2009["Date"].dt.month == j+1]["C2-Anemómetro 2 [m/s]"].std(ddof =1)
    
Desv = pd.DataFrame([Desv])
Desv
```

### Intensidad de Turbulencia

```python
Intensidad=Desv/Promedios
Intensidad
```

```python
Promedios.index = Promedios.index.where(Promedios.index != 0, 'Promedios')

resumen = Promedios.copy()

resumen.loc["Desv. Est"] = Desv.loc[0]
resumen.loc["Minimo"] = Minimos.loc[0]
resumen.loc["Maximo"] = Maximos.loc[0]
resumen.loc["Intensidad"] = Intensidad.loc[0]
resumen["Anual"] = resumen.mean(axis=1)
resumen = resumen.round(3)

resumen
```

Se puede ver que en promedio, los meses de Marzo, Abril y Febrero tienen las mayores velocidades entonces estos son los mejores meses

```python
fig, ax = plt.subplots(dpi = 250)
ax.plot(resumen.iloc[0][0:15].index ,resumen.iloc[0][0:15], marker = "o", color = "#41B06E")
ax.set_ylabel("Promedio")
ax.set_title("Velocidades (m/s)")
plt.subplots_adjust(hspace= 1, left= -2)
plt.show()
```
